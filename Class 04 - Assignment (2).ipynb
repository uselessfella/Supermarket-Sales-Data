
# Modeling

````python
X = df['Rating']
y = df['Total']
````

````python
from sklearn.model_selection import train_test_split
````

````python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.50,random_state=100)
````

````python
X_train.shape
````

````python
X_train.values.reshape(-1,1)
````

````python
X_train.values.reshape(-1,1).shape
````

````python
X_train = X_train.values.reshape(-1,1)
X_test = X_test.values.reshape(-1,1)
````

````python
from sklearn.linear_model import LinearRegression
````

````python
lr = LinearRegression()
````

````python
lr.fit(X_train,y_train)
````

````python
lr.predict([[2.9]])
````

````python
y_pred = lr.predict(X_test)
````

````python
X_test
````

````python
y_pred
````

````python
y_test
````

````python
len(y_test)
````

````python
x_range = [i for i in range(len(y_test))]
````

````python
x_range
````

````python
plt.scatter(x_range, y_test)
plt.scatter(x_range, y_pred)
plt.title('Prediction')
plt.xlabel('Rating')
plt.ylabel('Total')
plt.legend(["Real", "Predicted"])
plt.show()
````

````python
from sklearn.metrics import r2_score,mean_squared_error
````

````python
mean_squared_error(y_test,y_pred)
````

````python
mse = mean_squared_error(y_test,y_pred)
````

````python
rsq = r2_score(y_test,y_pred)
````

````python
print('mean squared error :',mse)
print('r square :',rsq)
````

````python
print('Intercept of the model:',lr.intercept_)
print('Coefficient of the line:',lr.coef_)
````
